## Supplementary Figure S1A: Comparison of measurements between barcode and extended primers


```python
# read expression data measured using extended primers
hela1 = pd.read_csv('ptreseq_raw_count/HELA-1_count.txt', sep = '\t')
hela1 = hela1.groupby('reporter').sum(numeric_only = True).reset_index()
hela1['CPM'] = hela1['count']*1000000/hela1['count'].sum()

hela2 = pd.read_csv('ptreseq_raw_count/HELA-2_count.txt', sep = '\t')
hela2 = hela2.groupby('reporter').sum(numeric_only = True).reset_index()
hela2['CPM'] = hela2['count']*1000000/hela1['count'].sum()

extended_rna_count = hela1.merge(hela2, on = 'reporter', suffixes = ('_1', '_2'))
extended_rna_count['CPM'] = extended_rna_count[['CPM_1', 'CPM_2']].mean(axis = 1)

# read input DNA count measured using extended primers
dna1 = pd.read_csv('ptreseq_raw_count/DNA-1_count.txt', sep = '\t')
dna1 = dna1.groupby('reporter').sum(numeric_only = True).reset_index()
dna1['CPM'] = dna1['count']*1000000/dna1['count'].sum()

dna2 = pd.read_csv('ptreseq_raw_count/DNA-2_count.txt', sep = '\t')
dna2 = dna2.groupby('reporter').sum(numeric_only = True).reset_index()
dna2['CPM'] = dna2['count']*1000000/dna2['count'].sum()

extended_dna_count = dna1.merge(dna2, on = 'reporter', suffixes = ('_1', '_2'))
extended_dna_count['CPM'] = extended_dna_count[['CPM_1', 'CPM_2']].mean(axis = 1)

# normalize RNA count by DNA count
extended_count = extended_rna_count[
    ['reporter', 'CPM']
].merge(
    extended_dna_count[['reporter', 'CPM']], 
    on = 'reporter', suffixes = ('_RNA', '_DNA')
)

extended_count['RNA/DNA'] = extended_count['CPM_RNA']/extended_count['CPM_DNA']

# Calculate BBBB median and normalize expression
BBBB = extended_count[extended_count['reporter'].str.contains('BBBB')]
median_exp = BBBB['RNA/DNA'].median()

# normalized by median of BBBB normalized count
extended_count['L2FC'] = np.log2(extended_count['RNA/DNA']/median_exp)

# calculate medioan
extended_count['RE_Identity'] = extended_count['reporter'].apply(lambda g: g.split('_')[0])
extended_count_median = extended_count.groupby('RE_Identity')['L2FC'].median()
```

```python
# Import barcode count from barcdoe primer
# read RNA expression data
ptreseq_rna_count = pd.read_excel('data/sup3_readscount_HeLa_total.xlsx')
ptreseq_rna_count = ptreseq_rna_count[ptreseq_rna_count['counts_RNA']>10]

# read plasmid
ptreseq_dna_count = pd.read_excel('data/sup2_readscount_plasmid.xlsx')
ptreseq_dna_count = ptreseq_dna_count.groupby(['Barcode'])['counts_Plasmid'].mean().reset_index()

# read barcode
bc = pd.read_csv('data/barcode_indexed.txt', sep = '\t', names = ['reporter', 'Barcode'])

# merge
ptreseq_count = ptreseq_rna_count.merge(ptreseq_dna_count, on = ['Barcode'], suffixes = ['_RNA', '_DNA'])
ptreseq_count = ptreseq_count.merge(bc, on = 'Barcode')

# rename some plasmids
rename_dict = {'Lpcx1': '7pcx1', 'Lpcx2': '7pcx2', 'Lpcx4': '7pcx4'}

def rename_f(p):
    if p in rename_dict.keys():
        return rename_dict[p]
    else:
        return p
    
ptreseq_count['RE_Identity'] = ptreseq_count['RE_Identity'].apply(lambda g: rename_f(g))

# normalize raw count by plasmid count
ptreseq_count['RNA/DNA'] = ptreseq_count['counts_RNA']/ptreseq_count['counts_Plasmid']

# Calculate BBBB median and normalize expression
BBBB = ptreseq_count[ptreseq_count['RE_Identity']=='BBBB']
median_exp = BBBB['RNA/DNA'].median()

# normalized by median of BBBB normalized count
ptreseq_count['L2FC'] = np.log2(ptreseq_count['RNA/DNA']/median_exp)

# calculate median
ptreseq_count_median = ptreseq_count.groupby('RE_Identity')['L2FC'].median()
```

```python
# merged measurements from barcode and extended primers
toplot = ptreseq_count_median.to_frame().merge(extended_count_median.to_frame(), on = 'RE_Identity', suffixes = ['_bc', '_ex'])
print(toplot[['L2FC_ex', 'L2FC_bc']].corr()) # Pearson correlation

f, ax = plt.subplots(figsize = (5, 5))
sns.regplot(
    data = toplot, 
    x = 'L2FC_ex', 
    y = 'L2FC_bc', 
    scatter_kws = {'color': 'black'},
    ax = ax
)

#f.savefig('Figures_ver3/Supplementary_Figure_1A_barcode_extended.pdf', transparent = True)
```

## Supplementary Figure 1D: piechart showing distribution of read assignments

```python
labels = ['full_length', 'spliced', 'RT_error', 'ambiguous', 'chimeric', 'cannot_find_barcode']
colors = ['royalblue', 'seagreen', 'peru', 'coral', 'darksalmon', 'grey']

def replace_class_name(c):
    if c in ['full_length', 'RT_error', 'ambiguous', 'chimeric', 'cannot_find_barcode']:
        return c
    else:
        return 'spliced'

def calculate_class_count(data):
    df = pd.read_csv(data, sep = '\t')
    df['class'] = df['class'].apply(lambda g: replace_class_name(g))
    df = df.groupby('class')['count'].sum()
    df = df.reindex(labels)
    
    return df
```

```python
# DNA plasmid

# rep 1
dna1 = calculate_class_count('ptreseq_raw_count/DNA-1_count.txt')
dna1['cannot_find_barcode'] = 1188607 # number of reads with unassinable bacrode
dna1_per = dna1/dna1.sum()
# rep 2
dna2 = calculate_class_count('ptreseq_raw_count/DNA-2_count.txt')
dna2['cannot_find_barcode'] = 318423 # number of reads with unassinable bacrode
dna2_per = dna2/dna2.sum()

# calculate average percentage
dna = (dna1_per + dna2_per)/2
dna = dna/dna.sum()
```

```python
# Superscript II

# rep 1
hela1 = calculate_class_count('ptreseq_raw_count/HELA-1_count.txt')
hela1['cannot_find_barcode'] = 3492501 # number of reads with unassinable bacrode
hela1_per = hela1/hela1.sum()
# rep 2
hela2 = calculate_class_count('ptreseq_raw_count/HELA-2_count.txt')
hela2['cannot_find_barcode'] = 1419736 # number of reads with unassinable bacrode
hela2_per = hela2/hela2.sum()

# calculate average percentage
hela = (hela1_per + hela2_per)/2
hela = hela/hela.sum()
```

```python
# Marathon

# rep 1
mara1 = calculate_class_count('ptreseq_raw_count/ETOH-1_count.txt')
mara1['cannot_find_barcode'] = 13623559 # number of reads with unassinable bacrode
mara1_per = mara1/mara1.sum()
# rep 2
mara2 = calculate_class_count('ptreseq_raw_count/ETOH-2_count.txt')
mara2['cannot_find_barcode'] = 10715961 # number of reads with unassinable bacrode
mara2_per = mara2/mara2.sum()

# calculate average percentage
mara = (mara1_per + mara2_per)/2
mara = mara/mara.sum()
```

```python
f, (ax1, ax2, ax3) = plt.subplots(ncols = 3, figsize = (10, 8))

# from left to right: DNA, HELA + SSII RT, HELA + Marathon RT
dna.plot.pie(y = 'count', ax = ax1, labels = labels, autopct='%1.1f%%', colors = colors)
hela.plot.pie(y = 'count', ax = ax2, labels = labels, autopct='%1.1f%%', colors = colors)
mara.plot.pie(y = 'count', ax = ax3, labels = labels, autopct='%1.1f%%', colors = colors)

for ax in (ax1, ax2, ax3):
    ax.set_ylabel('')
     
ax2.legend(bbox_to_anchor = (3.5, 1))

#f.savefig('piechart.pdf', transparent = True)
```

## Figure S1F: percent of RT recombination reads 

```python
def calculate_percent_RT_recomb(data):
    df = pd.read_csv(data, sep = '\t')
    total_count = df.groupby('class')['count'].sum()
    fraction = total_count/total_count.sum()
    
    try:
        return fraction['RT_error']
    except:
        return 0

# percent of RT recombination reads in DNA plasmids
dna1_RT_recomb = calculate_percent_RT_recomb('ptreseq_raw_count/DNA-1_count.txt')
dna2_RT_recomb = calculate_percent_RT_recomb('ptreseq_raw_count/DNA-2_count.txt')

# percent of RT recombination reads in Superscript II sample
hela1_RT_recomb = calculate_percent_RT_recomb('ptreseq_raw_count/HELA-1_count.txt')
hela2_RT_recomb = calculate_percent_RT_recomb('ptreseq_raw_count/HELA-2_count.txt')

# percent of RT recombination reads in Marathon sample
mara1_RT_recomb = calculate_percent_RT_recomb('ptreseq_raw_count/ETOH-1_count.txt')
mara2_RT_recomb = calculate_percent_RT_recomb('ptreseq_raw_count/ETOH-2_count.txt')
```

```python
label = ['DNA', 'DNA', 'HeLa (Marathon)', 'HeLa (Marathon)', 'HeLa (SSII)', 'HeLa (SSII)']
data = [dna1_RT_recomb, dna2_RT_recomb, mara1_RT_recomb, mara2_RT_recomb, hela1_RT_recomb, hela2_RT_recomb]

toplot = pd.DataFrame({'label': label, 'percent of RT recomb. reads': data})

f, ax = plt.subplots()

sns.barplot(data = toplot, x = 'label', y = 'percent of RT recomb. reads', ax = ax)

ax.set_yticks([0, 0.06, 0.12])
ax.set_yticklabels(['0%', '6%', '12%'])

#f.savefig('Figures_ver3/Supplementary_Figure_1F_RTrecomb_percentage.pdf')
```

## Figure 1H: reproducibility between biological replicates of HeLa (SSII) sample

```python
def process_raw_count(data):
    df = pd.read_csv(data, sep = '\t')
    
    # get sample name
    sample = os.path.basename(data).split('_')[0]
    df['sample'] = sample
    
    # remove artifact reads
    df = df[~df['class'].isin([
                'cannot_find_barcode', 
                'RT_error', 
                'chimeric', 
                'ambiguous', 
                'wrong length'
            ])
            ].reset_index(
        drop=True
    )
    
    # calculate total read count for each reporter
    df['total'] = df.groupby('reporter')['count'].transform('sum')
    # remove reporter with read count less than 10
    df = df[df['total']>=-1]
    # calculate fraction
    df['fraction'] = df['count']/df['total']
    # get regulatory element name
    df['RE'] = df['reporter'].apply(lambda g: g.split('_')[0])
    # rearrange column order 
    df = df[['sample', 'RE', 'reporter', 'class', 'count', 'total', 'fraction']]
    # rename column
    #df = df.rename(columns = {'count': f'count_{sample}', 'total': f'total_{sample}', 'fraction': f'fraction_{sample}'})
    
    return df

def compare_bio_rep(*args):
    df = pd.DataFrame(columns = ['RE'])
    
    for _df in args:
        FL = _df[_df['class']=='full_length']
        FL = FL.rename(columns = {'fraction': FL['sample'].values[0]})
        df = df.merge(FL[['RE', FL['sample'].values[0]]], on = 'RE', how = 'outer')

    df = df.fillna(0)
    df = df.groupby('RE').median()
    
    print(df.corr())
    return df
```

```python
# read data for HeLa cell
hela1 = process_raw_count('ptreseq_raw_count/HELA-1_count.txt')
hela2 = process_raw_count('ptreseq_raw_count/HELA-2_count.txt')

hela = compare_bio_rep(hela1, hela2)
```

```python
# reg plot
f, ax = plt.subplots(figsize = (5, 5))

sns.regplot(data = hela, x = 'HELA-1', y = 'HELA-2', ax = ax)

ax.set_xlim(-0.05, 1.05)
ax.set_ylim(-0.05, 1.05)

ax.set_xticks([0, 0.5, 1.0])
ax.set_xticklabels(['0%', '50%', '100%'])

ax.set_yticks([0, 0.5, 1.0])
ax.set_yticklabels(['0%', '50%', '100%'])

#f.savefig('Figures_ver3/Supplementary_Figure_1H_HeLa_bio_rep.pdf')
```

## Supplementary Figure 1I: mean absolute deviation between barcodes

```python
# combine 2 HELA replicates
hela_concat = pd.concat([hela1, hela2])
hela_concat_fl = hela_concat[hela_concat['class']=='full_length']
hela_concat_fl['spliced_fraction'] = 1 - hela_concat_fl['fraction']
hela_concat_fl_mad = hela_concat_fl.groupby('RE')['spliced_fraction'].mad() # calculate mean absolute deviation

# plot histogram
f, ax = plt.subplots()

sns.histplot(x = hela_concat_fl_mad, color = 'grey', bins = 50, ax = ax)

ax.set_xlabel('median absolute deviation of splicing efficiency across internal barcode replicates')
ax.set_ylabel('reporter count')

#f.savefig('Figures_ver3/Supplementary_Figure_1H_barcode_MAD.pdf')
```
